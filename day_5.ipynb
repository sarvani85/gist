## pie chart
    - dividing each piece of data (rows,columns)into seperate and visualize

```python
branchcount=[120,180,70,80,60]
branches=['CSE','ECE','EEE','MEC','CIVIL']
colors=['c','m','k','b','r']
plt.pie(branchcount,labels=branches,startangle=180,shadow=True,explode=(0.5,0,0,0,0))
plt.show()
```

```python
branchcount=[120,180,70,80,60]
branches=['CSE','ECE','EEE','MEC','CIVIL']
colors=['c','m','k','b','r']
plt.pie(branchcount,labels=branches,startangle=180,shadow=True,explode=(0.2,0,0,0,0))
plt.show()
```

```python
branchcount=[120,180,70,80,60]
branches=['CSE','ECE','EEE','MEC','CIVIL']
colors=['c','m','k','b','r']
plt.pie(branchcount,labels=branches,startangle=180,shadow=True,explode=(0.2,0.1,0,0,0))
plt.show()
```

```python
branchcount=[120,180,70,80,60]
branches=['CSE','ECE','EEE','MEC','CIVIL']
colors=['c','m','k','b','r']
plt.pie(branchcount,labels=branches,startangle=180,shadow=True,explode=(0.2,0.1,0,0,0),autopct='%1.1f%%')
plt.show()
```
